## 33. 「AのB」Permalink
2つの名詞が「の」で連結されている名詞句を抽出せよ

```python
sentence_list = []
result_list = []

#表層形\t品詞,品詞細分類1,品詞細分類2,品詞細分類3,活用型,活用形,原形,読み,発音

with open("neko.txt.mecab", "r") as f:
    lines = f.readlines()
    for word in lines:
        word_dic = {}
        surf = word.split('\t')
        if surf[0] == 'EOS\n':
            continue
        info = surf[1].split(',')
        word_dic["surface"] = surf[0]
        if len(info) < 8:
            word_dic["base"] = surf[0]
        else:
            word_dic["base"] = info[7]
        word_dic["pos"] = info[0]
        word_dic["pos1"] = info[1]
        sentence_list.append(word_dic)
        if surf[0] == "。":
            result_list.append(sentence_list)
            sentence_list = []

result_list
```

```python
o_list = []

for sentence in result_list:
    length = len(sentence)
    for i, word in enumerate(sentence):
        if word['pos'] == '名詞' and length > i + 2 and sentence[i+1]['surface'] == 'の' and sentence[i+2]['pos']== '名詞':
            o_list.append(word['surface'] + 'の' + sentence[i+2]['surface'])
o_list
```
