## 35. 単語の出現頻度Permalink
文章中に出現する単語とその出現頻度を求め，出現頻度の高い順に並べよ．

```python
sentence_list = []
result_list = []

#表層形\t品詞,品詞細分類1,品詞細分類2,品詞細分類3,活用型,活用形,原形,読み,発音

with open("neko.txt.mecab", "r") as f:
    lines = f.readlines()
    for word in lines:
        word_dic = {}
        surf = word.split('\t')
        if surf[0] == 'EOS\n':
            continue
        info = surf[1].split(',')
        word_dic["surface"] = surf[0]
        if len(info) < 8:
            word_dic["base"] = surf[0]
        else:
            word_dic["base"] = info[7]
        word_dic["pos"] = info[0]
        word_dic["pos1"] = info[1]
        sentence_list.append(word_dic)
        if surf[0] == "。":
            result_list.append(sentence_list)
            sentence_list = []
```

```python
import collections
tango_list = []

for sentence in result_list:
    for word in sentence:
        if word['pos'] != '空白' and word['pos'] != '補助記号' and word['pos'] != '記号' and word['pos'] != '助詞' and word['pos'] != '助動詞':
            tango_list.append(word['surface'])

c = collections.Counter(tango_list)
result = c.most_common()
result
```
