```python
import MeCab
import unidic
mecab = MeCab.Tagger()
with open("neko.txt", "r") as input, open("neko.txt.mecab", "w") as output:
    lines = input.readlines()
    for text in lines:
        result = mecab.parse(text)
        output.write(result)
```

```python
neko_dic = {}

with open("neko.txt.mecab", "r") as f:
    lines = f.readlines()
    for word in lines:
        tmp = word.split('\t')
        if tmp[0] == 'EOS\n':
            continue
        neko_dic[tmp[0]] = tmp[1]
          
neko_dic
```
